### 1. List and briefly define the possible states that define an instruction execution.

1.Fetch (IF): Se obtiene la instrucción a ser ejecutada desde memoria y se carga en el registro de instrucción.

2.Decode (ID): Se decodifica la instrucción para determinar el tipo de operación y los operandos necesarios, mientras se leen los registros para obtener los valores de los operandos.

3.Execute (EX): Se ejecuta la operación indicada por la instrucción dependiendo del tipo de operación. Si es una operación matemática, se ejecuta directamente, sino se calcula la dirección. 

4.Memory access (MEM): Si la operación requiere acceder a la memoria, se accede a un operando en memoria o lee desde una entrada o salida de datos.

5.Write back (WB): Finalmente, el resultado de la operación se escribe en un registro o memoria correspondiente.


### 2. What types of transfers must a computer’s interconnection structure (e.g., bus) support?

1.Bus de direcciones (Address bus): Este bus se utiliza para transmitir direcciones entre el procesador y la memoria. El procesador utiliza el bus de direcciones para enviar la dirección de memoria donde se encuentra la instrucción o el dato que desea acceder. Por ejemplo, cuando se ejecuta una instrucción, el procesador transmite la dirección de memoria de esa instrucción al registro de dirección de memoria (MAR) a través del bus de direcciones.

2.Bus de datos (Data bus): Este bus se utiliza para mover datos e instrucciones (el contenido de memoria o datos) entre el procesador y la memoria. El procesador utiliza el bus de datos para leer o escribir datos en la memoria. Por ejemplo, cuando se lee una instrucción o un dato de la memoria, el procesador utiliza el bus de datos para mover el contenido de la memoria al registro de datos de memoria (MDR) del procesador.

3.Bus de control (Control bus): Este bus se utiliza para transmitir señales de control entre el procesador y la memoria o los dispositivos de entrada/salida (I/O) para indicar la lectura o escritura de datos. Por ejemplo, cuando el procesador desea leer o escribir datos en la memoria, utiliza el bus de control para enviar una señal de control a la memoria indicando que se realizará una operación de lectura o escritura. También se utilizan señales de control para la transferencia de datos entre los dispositivos de entrada/salida y el procesador, indicando el inicio y fin de la transferencia de datos.
